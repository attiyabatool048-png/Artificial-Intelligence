```python
import pandas as pd
import numpy as np

from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn import metrics

uploaded = files.upload()

data = pd.read_csv('student_performance.csv')

x = data.iloc[:, 0:-1]
y = data.iloc[:, -1]

cat_columns = x.select_dtypes(include=['object']).columns
x[cat_columns] = x[cat_columns].apply(lambda col: pd.factorize(col)[0])

X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size=0.3, shuffle=False
)

BernNB = BernoulliNB()

BernNB.fit(X_train, Y_train)

Y_pred = BernNB.predict(X_test)

print("Accuracy:",
      metrics.accuracy_score(Y_test, Y_pred))

print("Precision:",
      metrics.precision_score(Y_test, Y_pred,
      average='weighted',
      labels=np.unique(Y_pred)))

print("Recall:",
      metrics.recall_score(Y_test, Y_pred,
      average='weighted',
      labels=np.unique(Y_pred)))

print("F1 Score:",
      metrics.f1_score(Y_test, Y_pred,
      average='weighted',
      labels=np.unique(Y_pred)))
```

```python
import pandas as pd
import numpy as np

from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder # Import LabelEncoder

uploaded = files.upload()

data = pd.read_csv('student_performance.csv')

# --- Start of data preprocessing to ensure non-negative features for MultinomialNB ---
# Fill null values for 'parent_education' if any, using the mode as done in Hng0ooqmFZQr
if 'parent_education' in data.columns:
    data['parent_education'] = data['parent_education'].fillna(data['parent_education'].mode()[0])

# Convert 'age' to int64, as done in Hng0ooqmFZQr
if 'age' in data.columns:
    data['age'] = data['age'].astype(np.int64)

# Drop 'student_id' column, as done in Hng0ooqmFZQr, before splitting x and y
if 'student_id' in data.columns:
    data.drop('student_id', axis=1, inplace=True)
# --- End of data preprocessing ---

# Separate features (x) and target (y)
x = data.iloc[:, :-1] # All columns except the last one for features
y = data.iloc[:, -1]  # The last column for the target

# Identify categorical columns (object type after initial preprocessing)
cat_columns = x.select_dtypes(include=['object']).columns

# Apply LabelEncoder to categorical columns
for col in cat_columns:
    # Fill any remaining NaNs in categorical columns before encoding.
    # LabelEncoder does not handle NaNs by default and will raise an error.
    # Filling with mode is a common strategy for categorical NaNs.
    if x[col].isnull().any():
        x[col] = x[col].fillna(x[col].mode()[0])

    le = LabelEncoder()
    x[col] = le.fit_transform(x[col])

# Ensure all numerical features are non-negative. This is a general safeguard.
# LabelEncoder already produces non-negative integers (0, 1, ...).
# However, if any other numerical column somehow contains negative values, this will shift them.
for col in x.columns:
    # Only apply this transformation to numerical columns (int/float)
    if x[col].dtype in [np.int64, np.float64, np.int32, np.float32]: # Added more specific integer/float types
        if x[col].min() < 0:
            x[col] = x[col] + abs(x[col].min()) # Shifts minimum value to 0

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size=0.3, shuffle=False
)

# --- Debugging step: Print minimum values in X_train ---
print("\nMinimum values in X_train before fitting MultinomialNB:")
print(X_train.min())
# --- End Debugging step ---

# Initialize and train the Multinomial Naive Bayes model
MultiNB = MultinomialNB()
MultiNB.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = MultiNB.predict(X_test)

# Evaluate the model
print("Accuracy:", metrics.accuracy_score(Y_test, Y_pred))
print("Precision:", metrics.precision_score(Y_test, Y_pred, average='weighted', labels=np.unique(Y_pred)))
print("Recall:", metrics.recall_score(Y_test, Y_pred, average='weighted', labels=np.unique(Y_pred)))
print("F1 Score:", metrics.f1_score(Y_test, Y_pred, average='weighted', labels=np.unique(Y_pred)))
```
